```python
import pandas as pd
import sys,os

path = os.getcwd()
path = path.replace('data_process','')
print(path)
```

```python
df = pd.read_csv(path + '//data0819-0918-校正.csv', encoding='utf-8')

df = df.drop(columns=['模型判断结果（政科）', '模型判断结果（中科院）'])
df = df.rename(columns={'校正（郭）': '政务分类'})

df.to_csv(path + '//data0819-0918-校正-cleaned.csv', index=False, encoding='utf-8-sig')
```

```python
df_confusing = pd.read_csv(path + '//data0819-0918-confusing.csv')

mask = df_confusing['模型判断结果（政科）'] == df_confusing['模型判断结果（中科院）']
df_confusing_filtered = df_confusing[mask].copy()

df_confusing_filtered['政务分类'] = df_confusing_filtered['模型判断结果（政科）']
df_confusing_filtered = df_confusing_filtered.drop(columns=['模型判断结果（政科）', '模型判断结果（中科院）'])

df_cleaned = pd.read_csv(path + '//data0819-0918-校正-cleaned.csv')

final_df = pd.concat([df_cleaned, df_confusing_filtered], axis=0)
final_df = final_df.drop_duplicates()

final_df.to_csv(path + '//data0819-0918-final.csv', index=False, encoding='utf-8-sig')
```

```python
df = pd.read_csv(path + '//data0319-0419.csv')
df['政务分类'] = df['模型判断结果']
df = df.drop(columns=['模型判断结果'])
df.to_csv(path + 'data0319-0419_with_gov_class.csv', index=False)

df = pd.read_csv(path + '//data0420-0518.csv')
df['政务分类'] = df['模型判断结果']
df = df.drop(columns=['模型判断结果'])
df.to_csv(path + 'data0420-0518_with_gov_class.csv', index=False)

df = pd.read_csv(path + '//data0519-0618.csv')
df['政务分类'] = df['模型判断结果']
df = df.drop(columns=['模型判断结果','Unnamed: 31','Unnamed: 32','Unnamed: 33','Unnamed: 34','Unnamed: 35','Unnamed: 36','Unnamed: 37','Unnamed: 38','Unnamed: 39','Unnamed: 40','Unnamed: 41'])
df.to_csv(path + 'data0519-0618_with_gov_class.csv', index=False)
```

```python
files = [path + '//data0319-0419_with_gov_class.csv',path + '//data0420-0518_with_gov_class.csv',path + '//data0519-0618_with_gov_class.csv',path + '//data0819-0918-final.csv']
dfs = [pd.read_csv(f) for f in files]

merged_df = pd.concat(dfs, ignore_index=True)

cleaned_df = merged_df.drop_duplicates()

cleaned_df['创建时间'] = pd.to_datetime(cleaned_df['创建时间'], format='mixed')

sorted_df = cleaned_df.sort_values(by='创建时间')

sorted_df = sorted_df.drop(columns=['序号'], errors='ignore')

sorted_df.insert(0, '序号', range(1, len(sorted_df)+1))

sorted_df.to_csv(path + '//merged_data.csv', index=False)
print("合并完成，结果已保存为merged_data.csv")
```
